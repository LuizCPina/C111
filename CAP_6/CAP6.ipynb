```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

```

Questao 1)

```python
dataset = np.loadtxt("paises.csv",delimiter=  ';',dtype=str, encoding='utf-8')
datasetAmerica = dataset[dataset[:,1] == 'NORTHERN AMERICA                   ']
birthDatasetAmerica = datasetAmerica[:, 15].astype(float)
deathDatasetAmerica = datasetAmerica[:, 16].astype(float)
print(birthDatasetAmerica)
print(deathDatasetAmerica)
plt.xlabel("Paises")
plt.ylabel("Taxaa de Mortalidade/Natalidade")
plt.title("Taxa de Natalidade(rosa) e Mortalidade(azul) - América do Norte")
plt.plot( datasetAmerica[:,0],birthDatasetAmerica, "m:*",  datasetAmerica[:,0],deathDatasetAmerica, "c:*")
```

Questao 2)

```python
dataframe = pd.read_csv("space.csv", delimiter=';', encoding='utf-8')
dataframeUSA = dataframe[dataframe["Location"].str.contains("USA")]["Company Name"].unique()  
print(dataframeUSA)
dataframeChina = dataframe[dataframe["Location"].str.contains("China")]["Company Name"].unique()
print(dataframeChina)

countries = ["USA", "China"]    
valores = [len(dataframeUSA), len(dataframeChina)]

plt.figure(figsize=(10,5))
plt.bar(countries, valores, color=['green','purple'])
plt.title("Número de empresas espaciais por país")
plt.xlabel("Países")
plt.ylabel("Número de empresas espaciais")
```

Questao 3)

```python
dataframeRoscT = dataframe[dataframe["Company Name"]=="Roscosmos"]["Status Mission"]
dataframeRoscS = dataframeRoscT[dataframeRoscT == "Success"]
dataframeRoscF = dataframeRoscT[dataframeRoscT == "Failure"]

plt.figure(figsize=(6,6))
plt.pie([len(dataframeRoscS), len(dataframeRoscF)], labels=["Sucesso", "Falha"], autopct='%1.1f%%', startangle=60, colors=['magenta', 'blue'])
plt.title("Taxa de sucesso e falha da Roscosmos")
```
